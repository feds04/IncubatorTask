```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('d:\\incubator\\PARTD_PRESCRIBER_PUF_NPI_DRUG_13.csv')
```

```python
df3 = df.groupby('GENERIC_NAME')['TOTAL_DRUG_COST'].sum()
df3_sorted=pd.Series(df3.sort(inplace=False,ascending=False))
```

```python
is_generic=df[df['DRUG_NAME']==df['GENERIC_NAME']]
generic_by_state=is_generic[u'NPPES_PROVIDER_STATE'].value_counts()
total_by_state=df[u'NPPES_PROVIDER_STATE'].value_counts()
#combined=pd.DataFrame[generic_by_state,generic_total]
generic_combined=pd.concat([generic_by_state/total_by_state,1-generic_by_state/total_by_state], axis=1)
generic_combined.columns=['generic', 'brand']
generic_combined
fig=generic_combined.plot(kind='bar', stacked=True, figsize=(12, 9),color=['b','r'], title='Fraction of brand name vs generic prescriptions per state')

plt.savefig("distrib_by_state.png")
```

```python
df3 = df.groupby(u'SPECIALTY_DESC')['TOTAL_DRUG_COST'].sum()
df3
top=pd.Series(df3.sort(inplace=False,ascending=False))[:5]
top.name=''
top['Other']=df3.sum()-top.sum()
top.plot(kind='pie', figsize=(6, 6)).set_title('Costs by specialty type')
plt.savefig("distrib_by_specialty.png")
```

```python
top_costs=[]
for i in range(df3_sorted.count()):
    top_costs=top_costs+[df3_sorted[0:i].sum()]

plt.xlabel('Number of the most expensive drugs')
plt.ylabel('Contribution to the total expenses')
plt.title('Cost plot')
plt.plot(np.array(top_costs)/df3_sorted.sum())
plt.savefig("top_contribution.png")
```
